```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
sample=[60,100,150,200,300,500]
bin=[10,15,18,21,24,30]
for i in range(6):
    N=np.random.uniform(0,1,sample[i])
    for j in range(6):
        
        plt.hist(N, bins=bin[j], density=1)
        plt.title(label="A")
        plt.show()
```
